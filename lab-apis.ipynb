```python
artists = ["Los Fabulosos Cadillacs", "Manu Chao", "Muchachito Bombo Infierno"]
```

```python
def get_top_tracks(artist_name, sp):
    # Search for the artist by name
    result = sp.search(q=artist_name, type='artist', limit=1)
    artist_id = result['artists']['items'][0]['id']
    
    # Fetch the top tracks for the artist
    top_tracks = sp.artist_top_tracks(artist_id)
    
    # Get the names of the first 5 top tracks
    return [track['name'] for track in top_tracks['tracks'][:5]]
```

```python
def find_related_artists(artist_name, sp):
    # Search for the artist by name
    result = sp.search(q=artist_name, type='artist', limit=1)
    artist_id = result['artists']['items'][0]['id']
    
    # Fetch related artists
    related_artists = sp.artist_related_artists(artist_id)
    
    # Get the names of the first 5 related artists
    return [artist['name'] for artist in related_artists['artists'][:5]]
```

```python
top_tracks_list = []
for artist in artists:
    top_tracks = get_top_tracks(artist, sp)
    top_tracks_list.extend(top_tracks)
```

```python
related_artists_list = []
related_top_tracks_list = []

for artist in artists:
    related_artists = find_related_artists(artist, sp)
    related_artists_list.extend(related_artists)
    
    for related_artist in related_artists:
        related_top_tracks = get_top_tracks(related_artist, sp)
        related_top_tracks_list.extend(related_top_tracks)
```

```python
playlist_tracks = top_tracks_list + related_top_tracks_list

print("Top tracks of favorite artists:")
print(top_tracks_list)
print("\nTop tracks of related artists:")
print(related_top_tracks_list)
print("\nCombined Playlist:")
print(playlist_tracks)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Initialize SpotiPy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID,
                                                           client_secret=CLIENT_SECRET))

# List of favorite artists
artists = ["Los Fabulosos Cadillacs", "Manu Chao", "Muchachito Bombo Infierno"]

def get_top_tracks(artist_name, sp):
    result = sp.search(q=artist_name, type='artist', limit=1)
    artist_id = result['artists']['items'][0]['id']
    top_tracks = sp.artist_top_tracks(artist_id)
    return [track['name'] for track in top_tracks['tracks'][:5]]

def find_related_artists(artist_name, sp):
    result = sp.search(q=artist_name, type='artist', limit=1)
    artist_id = result['artists']['items'][0]['id']
    related_artists = sp.artist_related_artists(artist_id)
    return [artist['name'] for artist in related_artists['artists'][:5]]

# Get top tracks for favorite artists
top_tracks_list = []
for artist in artists:
    top_tracks = get_top_tracks(artist, sp)
    top_tracks_list.extend(top_tracks)

# Find related artists and their top tracks
related_artists_list = []
related_top_tracks_list = []

for artist in artists:
    related_artists = find_related_artists(artist, sp)
    related_artists_list.extend(related_artists)
    
    for related_artist in related_artists:
        related_top_tracks = get_top_tracks(related_artist, sp)
        related_top_tracks_list.extend(related_top_tracks)

# Combine results into a playlist
playlist_tracks = top_tracks_list + related_top_tracks_list

# Print results
print("Top tracks of favorite artists:")
print(top_tracks_list)
print("\nTop tracks of related artists:")
print(related_top_tracks_list)
print("\nCombined Playlist:")
print(playlist_tracks)
```

## Playlists

The `sp.featured_playlists()` method in `spotipy` fetches a list of Spotify's featured playlists at a given moment. These are curated playlists that Spotify often highlights on the platform's homepage. The method provides a snapshot of the playlists that are being promoted or featured by Spotify at the time of the request.

Once you've fetched the featured playlists, you can extract their IDs (and other details).

```python
sp.featured_playlists() # We get a playlist id of a playlist we like
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Initialize SpotiPy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID,
                                                           client_secret=CLIENT_SECRET))

def get_featured_playlists(sp):
    # Fetch the featured playlists
    playlists = sp.featured_playlists()
    
    # Extract playlist details
    featured_playlists = []
    for playlist in playlists['playlists']['items']:
        playlist_info = {
            'id': playlist['id'],
            'name': playlist['name'],
            'description': playlist.get('description', 'No description available'),
            'external_url': playlist['external_urls']['spotify']
        }
        featured_playlists.append(playlist_info)
    
    return featured_playlists

# Get and print featured playlists
featured_playlists = get_featured_playlists(sp)

print("Featured Playlists:")
for playlist in featured_playlists:
    print(f"ID: {playlist['id']}")
    print(f"Name: {playlist['name']}")
    print(f"Description: {playlist['description']}")
    print(f"URL: {playlist['external_url']}")
    print()
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Initialize SpotiPy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID,
                                                           client_secret=CLIENT_SECRET))

def get_playlist_tracks(playlist_id, sp):
    # Fetch playlist tracks
    tracks = sp.playlist_tracks(playlist_id)
    
    # Extract track details
    track_list = []
    for item in tracks['items']:
        track = item['track']
        track_info = {
            'name': track['name'],
            'artists': [artist['name'] for artist in track['artists']],
            'album': track['album']['name'],
            'external_url': track['external_urls']['spotify']
        }
        track_list.append(track_info)
    
    return track_list

# Playlist ID
playlist_id = '37i9dQZF1DXd9zR7tdziuQ'

# Get and print tracks from the playlist
playlist_tracks = get_playlist_tracks(playlist_id, sp)

print("Tracks in Playlist:")
for track in playlist_tracks:
    print(f"Name: {track['name']}")
    print(f"Artists: {', '.join(track['artists'])}")
    print(f"Album: {track['album']}")
    print(f"URL: {track['external_url']}")
    print()
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Initialize SpotiPy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID,
                                                           client_secret=CLIENT_SECRET))

def get_artists_from_playlist(playlist_id, sp):
    # Fetch playlist tracks
    tracks = sp.playlist_tracks(playlist_id)
    
    # Set to store unique artist names
    artists_set = set()
    
    # Extract artists from each track
    for item in tracks['items']:
        track = item['track']
        for artist in track['artists']:
            artists_set.add(artist['name'])
    
    return list(artists_set)

# Playlist ID
playlist_id = '37i9dQZF1DXd9zR7tdziuQ'

# Get and print artists from the playlist
artists = get_artists_from_playlist(playlist_id, sp)

print("Artists in Playlist:")
for artist in artists:
    print(artist)
```

## Exercise 2: Unraveling the World of Playlists


1. **Featured Exploration**: 
   - Fetch the list of Spotify's current featured playlists. 
   - Extract and display the names and IDs of the top 5 featured playlists.
   
2. **Deep Dive**:
   - Choose any one of the top 5 featured playlists (you can choose the one you personally find most interesting or simply pick one randomly).
   - Fetch and display its name, description, and total track count.

3. **Track-tastic**:
   - Extract and display the names of the first 10 tracks in the chosen playlist.

4. **Artistic Flair**:
   - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.
   - For example: `{"TrackName1": ["Artist1", "Artist2"], "TrackName2": ["Artist3"]}`
   

```python
# Your answer here

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Initialize SpotiPy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID,
                                                           client_secret=CLIENT_SECRET))

def get_featured_playlists(sp):
    # Fetch the featured playlists
    playlists = sp.featured_playlists()
    # Extract details of the top 5 featured playlists
    featured_playlists = []
    for playlist in playlists['playlists']['items'][:5]:
        playlist_info = {
            'id': playlist['id'],
            'name': playlist['name']
        }
        featured_playlists.append(playlist_info)
    return featured_playlists

def get_playlist_details(playlist_id, sp):
    # Fetch playlist details
    playlist = sp.playlist(playlist_id)
    return {
        'name': playlist['name'],
        'description': playlist.get('description', 'No description available'),
        'tracks_count': playlist['tracks']['total']
    }

def get_playlist_tracks(playlist_id, sp):
    # Fetch playlist tracks
    tracks = sp.playlist_tracks(playlist_id)
    # Extract the names of the first 10 tracks
    track_list = []
    for item in tracks['items'][:10]:
        track = item['track']
        track_list.append(track['name'])
    return track_list

def get_artists_from_playlist_tracks(playlist_id, sp):
    # Fetch playlist tracks
    tracks = sp.playlist_tracks(playlist_id)
    # Extract artists for the first 10 tracks
    artists_dict = {}
    for item in tracks['items'][:10]:
        track = item['track']
        track_name = track['name']
        artists_list = [artist['name'] for artist in track['artists']]
        artists_dict[track_name] = artists_list
    return artists_dict

# 1. Featured Exploration
featured_playlists = get_featured_playlists(sp)
print("Top 5 Featured Playlists:")
for playlist in featured_playlists:
    print(f"Name: {playlist['name']}, ID: {playlist['id']}")

# Choose one playlist from the top 5 (for example, the first one)
chosen_playlist_id = featured_playlists[0]['id']

# 2. Deep Dive
playlist_details = get_playlist_details(chosen_playlist_id, sp)
print("\nChosen Playlist Details:")
print(f"Name: {playlist_details['name']}")
print(f"Description: {playlist_details['description']}")
print(f"Total Tracks: {playlist_details['tracks_count']}")

# 3. Track-tastic
playlist_tracks = get_playlist_tracks(chosen_playlist_id, sp)
print("\nFirst 10 Tracks in the Playlist:")
for track in playlist_tracks:
    print(track)

# 4. Artistic Flair
artists_dict = get_artists_from_playlist_tracks(chosen_playlist_id, sp)
print("\nArtists for the First 10 Tracks:")
for track_name, artists in artists_dict.items():
    print(f"{track_name}: {', '.join(artists)}")
```
